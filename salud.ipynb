```python
import pandas as pd

# Datos extraídos de la imagen
data = {
    "Temperatura": [36.5, 37.2, 36.8, 37.5, 36.9, 37.0, 35.3, 37.1, 34.9, 38.1, 36.7, 36.5, 37.22],
    "Presión": [120, 140, 130, 150, 125, 145, 129, 141, 153, 151, 125, 158, 160],
    "Comorbilidad": ["Obesidad", "Obesidad", "Ansiedad", "Asma", "Asma", "Arritmia", "Ansiedad", "Obesidad", "Ansiedad", "Obesidad", "Ansiedad", "Asma", "Arritmia"],
    "Clasificación": ["S", "E", "S", "E", "S", "E", "S", "E", "S", "E", "S", "E", "E"]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Filtrar datos donde la clasificación es "S"
df_s = df[df["Clasificación"] == "S"]

# Calcular la media de la temperatura en estos casos
mu_pre_s = df_s["Presión"].mean()
# Calcular la media de la temperatura en estos casos
mu_temp_s = df_s["Temperatura"].mean()


# Calcular la varianza muestral de la temperatura para la clasificación "S"

var_pre_s =df_s["Presión"].var()

var_temp_s = df_s["Temperatura"].var()



# Filtrar datos donde la clasificación es "S"
df_e = df[df["Clasificación"] == "E"]

# Calcular la media de la temperatura en estos casos
mu_pre_e = df_e["Presión"].mean()
# Calcular la media de la temperatura en estos casos
mu_temp_e = df_e["Temperatura"].mean()


# Calcular la varianza muestral de la temperatura para la clasificación "S"

var_pre_e =df_e["Presión"].var()

var_temp_e = df_e["Temperatura"].var()
```

```python
# Filtrar por clase S y E
df_s = df[df["Clasificación"] == "S"]
df_e = df[df["Clasificación"] == "E"]

# Calcular media y varianza para temperatura y presión en cada clase
stats = {
    "Clase": ["S", "E"],
    "Media Temp (μ)": [df_s["Temperatura"].mean(), df_e["Temperatura"].mean()],
    "Varianza Temp (σ^2)": [df_s["Temperatura"].var(), df_e["Temperatura"].var()],
    "Media Presión (μ)": [df_s["Presión"].mean(), df_e["Presión"].mean()],
    "Varianza Presión (σ^2)": [df_s["Presión"].var(), df_e["Presión"].var()],
}

# Crear DataFrame de estadísticas
df_stats = pd.DataFrame(stats)

# Calcular P(x3 | y) para cada clase
prob_x3_given_s = df_s["Comorbilidad"].value_counts(normalize=True)
prob_x3_given_e = df_e["Comorbilidad"].value_counts(normalize=True)
```

```python

# Valores del vector a clasificar
x1_new, x2_new, x3_new = 35.9, 143, "Obesidad"

# Media y desviación estándar para cada clase
mu_temp_s, sigma2_temp_s = df_s["Temperatura"].mean(), df_s["Temperatura"].var()
mu_temp_e, sigma2_temp_e = df_e["Temperatura"].mean(), df_e["Temperatura"].var()

mu_pres_s, sigma2_pres_s = df_s["Presión"].mean(), df_s["Presión"].var()
mu_pres_e, sigma2_pres_e = df_e["Presión"].mean(), df_e["Presión"].var()

# Cálculo de las probabilidades usando la distribución normal
p_x1_given_s = norm.pdf(x1_new, loc=mu_temp_s, scale=np.sqrt(sigma2_temp_s))
p_x2_given_s = norm.pdf(x2_new, loc=mu_pres_s, scale=np.sqrt(sigma2_pres_s))

p_x1_given_e = norm.pdf(x1_new, loc=mu_temp_e, scale=np.sqrt(sigma2_temp_e))
p_x2_given_e = norm.pdf(x2_new, loc=mu_pres_e, scale=np.sqrt(sigma2_pres_e))

# Probabilidad de comorbilidad en cada clase
p_x3_given_s = prob_x3_given_s.get(x3_new, 0)  # Si no existe, ponemos 0
p_x3_given_e = prob_x3_given_e.get(x3_new, 0)

# Probabilidades a priori
p_s = len(df_s) / len(df)
p_e = len(df_e) / len(df)

# Calcular las probabilidades totales
p_x_given_s = p_x1_given_s * p_x2_given_s * p_x3_given_s
p_x_given_e = p_x1_given_e * p_x2_given_e * p_x3_given_e

p_s_given_x = p_x_given_s * p_s
p_e_given_x = p_x_given_e * p_e

# Determinar la clase
classification = "S" if p_s_given_x > p_e_given_x else "E"

p_s_given_x, p_e_given_x, classification
```
